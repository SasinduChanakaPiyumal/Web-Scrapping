## Find_all()

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.text,"lxml")
prices = soup.find_all("h4",class_ = "float-end price card-title pull-right")
print(prices)
```

```python
print(len(prices))
```

```python
price = []
for i in prices:
    print(i.text)
    price.append(i.text)
```

```python
desc = soup.find_all("p",class_ = "description card-text")
print(desc)
```

```python
desc_list = []
for j in desc:
    print(j.text)
    print("**********************************")
    desc_list.append(j.text)
```

```python
import re

data = soup.find_all(string = re.compile("Prestigio"))
```

```python
print(data)
```

```python
import pandas as pd
```

```python
title = soup.find_all("a",class_ = "title")
print(title)
```

```python
product_name =[]
for i in title:
    name = i.text
    product_name.append(name)
print(product_name)    
```

```python
print(price)
```

```python
print(desc_list)
```

```python
review = soup.find_all("p",class_ = "float-end review-count")
#print(review)
review_list = []
for k in review:
    review_list.append(k.text)
print(review_list)
```

```python
df = pd.DataFrame({"Product Name":product_name,"Description":desc_list,"Price":price,"Reviews":review_list})
print(df)
```
